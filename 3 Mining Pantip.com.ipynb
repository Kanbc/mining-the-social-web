```python
import scrapy
from scrapy.crawler import CrawlerProcess

# point Scrapy to Pantip.com
class PantipSpider(scrapy.Spider):
    name = "pantip_spider"
    start_urls = ['https://pantip.com/#trend']
    
    def parse(self, response):
        SET_SELECTOR = '.post-pick-nothumb'
        for topic in response.css(SET_SELECTOR):
            
            NAME_SELECTOR = 'a .post-pick-title ::text'
            DESC_SELECTOR = 'a .post-pick-desc ::text'
            BY_SELECTOR = 'div span ::text'
            
            yield {
                'title': topic.css(NAME_SELECTOR).extract_first(),
                'desc': topic.css(DESC_SELECTOR).extract_first(),
                'by': topic.css(BY_SELECTOR).extract_first(),
            }

# run Scrapy
process = CrawlerProcess()
process.crawl(PantipSpider)
process.start() # the script will block here until the crawling is finished
```
